```python
count = 1
for key in newdfCols:
    subnewdf = newdf[newdf['phases.name'] == key]
    subnewdfCols = subnewdf['parameter'].value_counts().keys()
    for subkey in subnewdfCols:
#         print subkey
        temp = subnewdf[subnewdf['parameter']==subkey]
        
        plt.hist(temp['numeric_val'], 30);
#         sns.distplot(temp['numeric_val'], bins=50, kde=False,);
        plt.title(key+'---'+subkey)
        plt.savefig(key[:len(key)-7]+'-'+subkey+'.png', dpi=300)
        plt.clf()
        print count, '-->', key[:len(key)-7]+'-'+subkey, np.mean(temp['numeric_val']),'-->', np.std(temp['numeric_val']), '-->', np.max(temp['numeric_val']), '-->', np.min(temp['numeric_val'])
        count += 1
    
```
