```python
df = pd.read_csv(r'/Users/fanchang/Desktop/GRR_example2.csv')
df.head(5)
```

```python
dfPartMean = df.groupby('SN').mean()
dfMean = df.groupby(['Operator', 'SN']).mean()
Xbar = dfMean.groupby('Operator')['Metrics'].mean()
dfMax = df.groupby(['Operator', 'SN']).max()
dfMin = df.groupby(['Operator', 'SN']).min()
Range = dfMax.Metrics-dfMin.Metrics
Rbar = Range.groupby('Operator').mean()
Rbarbar = Rbar.mean()
Xbardiff = Xbar.max()-Xbar.min()
RP = (dfPartMean.max()-dfPartMean.min()).Metrics
D4 = 2.58
UCL = D4*Rbarbar
K1 = 0.5908
EV = Rbarbar*K1
numOfparts = 10
numOftrials = 3
K2 = 0.5231
AV = np.sqrt((Xbardiff*K2)**2-EV**2/(numOfparts*numOftrials))
GRR = np.sqrt(EV**2+AV**2)
K3 = 0.3146
PV = RP*K3
TV = np.sqrt(GRR**2+PV**2)
USL = 1
LSL = .55
```
