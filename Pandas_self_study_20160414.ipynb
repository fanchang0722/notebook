```python
s = pd.Series([0.0, 0.1, 0.2, 0.3, 0.4],
                 ['V', 'W', 'X', 'Y', 'Z'])
heatmap_data = pd.DataFrame({'A' : s + 0.0,
                                'B' : s + 0.1,
                                'C' : s + 0.2,
                                'D' : s + 0.3,
                                'E' : s + 0.4,
                                'F' : s + 0.5,
                                'G' : s + 0.6
                        })
heatmap_data
```
