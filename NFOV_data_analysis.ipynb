```python
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
import seaborn as sns
%pylab inline

def YieldProjection(df):
    NFOVPass=df[df.DUT_Status=='PASS']
    SNPass = NFOVPass.dut_id.unique()
    # failed SN
    NFOVFail=df[df.status <>'PASS']
    SNFail = NFOVFail.dut_id.unique()
    count = 0
#     print "failed SN is: "
    trueFail = []
    for SN in SNFail:
        if SN not in SNPass:
#             print '{:>22}'.format(SN,)
            trueFail.append(SN)
            count +=1
    
    print 'The total count of failed unit is {}'.format(count)
    print 'The total count of unit is {}'.format((len(SNPass)+count))
    print "The first time yield is {}%". format (round((len(SNPass)-len(SNFail)+count)*100.0/(len(SNPass)+count), 3))
    print "The final yield is {}%". format (round(100-count*100.0/(len(SNPass)+count), 3))
    return trueFail
```
