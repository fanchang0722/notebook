```python
import numpy as np
import matplotlib
import matplotlib.pylab as plt
import seaborn as sns
```

```python
p1 = np.array([1, 0])
p2 = np.array([3./2, 0])
p3 = np.array([2, -1])
plt.plot([p1[0], p3[0]], [p1[1], p3[1]],  '-o', lw=3, markersize=8)
plt.plot([p2[0], p3[0]], [p2[1], p3[1]],  '-x', lw=3, markersize=8)
plt.grid()
```
