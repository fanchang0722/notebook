```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%pylab inline
```

```python
folder = r'/Users/fanchang/Desktop/A70/'
subfolder = ['20180320', '20180321']

for idx in range(len(subfolder)):
#     print subfolder[idx]
    fileNames = os.listdir(os.path.join(folder, subfolder[idx]))
    count = 0
    fout = open(os.path.join(folder, subfolder[idx]+'.csv'), 'w')
#     fout.writelines(
#     print os.path.join(folder, subfolder[idx]+'.csv')
    for fileidx in range(len(fileNames)):
        fileName = fileNames[fileidx]
        if fileName.endswith('result.csv') and os.path.getsize(os.path.join(os.path.join(folder, subfolder[idx]), fileName))> 80:
            fp = open(os.path.join(os.path.join(folder, subfolder[idx]), fileName), 'r')
            fp.readline()
            temp = fp.readline().rstrip('\r\n')
            fout.writelines(temp + '\r\n')
#             print temp
            fp.close()
#             print 4*' ' + fileName
            count += 1
    fout.close()
    print "there is %s results in %s" % (count, subfolder[idx])
```
