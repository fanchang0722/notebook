```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.linspace(0, 0.5)
y = np.sinc(x)
# plt.figure(1, figsize = (11, 8.5))
# h, w = 3, 1
pylab.figure(figsize=(12,9))
# pylab.subplots_adjust(hspace=.7)

# pylab.subplot(h,w,1)
plt.plot(x, y)
plt.grid(True)
plt.xlim([0, 0.5])
plt.ylim([0.6, 1])
pylab.savefig("SIG.png",dpi=200)
plt.show()
```
